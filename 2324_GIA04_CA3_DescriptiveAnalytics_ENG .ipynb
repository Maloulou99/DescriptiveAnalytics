# Evaluation of sixth grade of primary education

## Context
The dataset *educacio.csv* contains the results obtained by sixth grade students in the assessment of basic skills and knowledge at the end of primary education since 2018.

We want to analyze this data set to provide information about the profile of 6th grade students and their annual evolution.

**Description of the columns**
* ANY: year in which the assessment took place.
* PCAT: overall weighted score of linguistic competence in the Catalan language.
* PCAST: weighted global score of linguistic competence in the Spanish language.
* PMAT: overall weighted math proficiency score.
* PANG: weighted global score of linguistic competence in the English language.
* PMED: global score of the competence associated with the area of knowledge of the natural environment.
* GENERE: gender of the student taking the assessment.
* ANY_NAIXEMENT: year of birth of the student taking the assessment.
* AREA_TERRITORIAL: region where the center of the student appearing for the assessment is located.
* NATURALESA: determines whether the student's center is public or private.
* HABITAT: municipalities by population sections.

## 1. Students analysis

1\.1. Show in different graphs the number of students per year, per year and nature, per year and gender, and per year and habitat.<br>
1\.2. Discuss your conclusions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('descriptiveAnalytics.csv')

fig, axes = plt.subplots(4, 1, figsize=(10, 20))

# Graph for number of students per year
sns.countplot(x='ANY', data=df, ax=axes[0], palette='viridis')
axes[0].set_title('Number of Students per Year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of Students')

# Graph for number of students per year and nature
sns.countplot(x='ANY', hue='NATURALESA', data=df, ax=axes[1], palette='muted')
axes[1].set_title('Number of Students per Year and Nature')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Number of Students')

# Graph for number of students per year and gender
sns.countplot(x='ANY', hue='GENERE', data=df, ax=axes[2], palette='pastel')
axes[2].set_title('Number of Students per Year and Gender')
axes[2].set_xlabel('Year')
axes[2].set_ylabel('Number of Students')

# Graph for number of students per year and habitat
sns.countplot(x='ANY', hue='HABITAT', data=df, ax=axes[3], palette='Set2')
axes[3].set_title('Number of Students per Year and Habitat')
axes[3].set_xlabel('Year')
axes[3].set_ylabel('Number of Students')

plt.tight_layout()
plt.show()

#Conlusion for every graph
print("In the 'Number of Students per Year' graph, we observe a relatively stable number of students in 2018 and 2019,") 
print("followed by a decrease in 2021. This decline in student enrollment may be attributed to the COVID-19 pandemic, ")
print("which necessitated remote learning via platforms like Zoom instead of in-person attendance at educational institutions.")
print("The challenges posed by remote learning conditions likely impacted some students' ")  
print("ability to effectively learn.This trend is mirrored in the subsequent graphs depicting the number of")     
print("students per year and categorized by nature, gender, and habitat. In each case, there is a notable decrease in")   
print("enrollment in 2021, potentially reflecting the broader societal impacts of the pandemic. However, we see a ")        
print("rebound in 2022, coinciding with the gradual reopening of various sectors and institutions.Overall,")          
print("these observations suggest that the fluctuations in student enrollment align closely with the disruptive")         
print("effects of the COVID-19 pandemic, underscoring the importance of considering external factors when analyzing")   
print("educational trends.")       
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('descriptiveAnalytics.csv')

# Calculate the average of the 5 grades
df['Average_Grade'] = df[['PCAT', 'PCAST', 'PANG', 'PMAT', 'PMED']].mean(axis=1)

# 3.1. Boxplots per year
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['ANY', 'PCAT', 'PCAST', 'PANG', 'PMAT', 'PMED', 'Average_Grade']], x='ANY')
plt.title('Boxplots of Tests and Average per Year')
plt.xlabel('Year')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()

# 3.2. Boxplots by gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['GENERE', 'PCAT', 'PCAST', 'PANG', 'PMAT', 'PMED', 'Average_Grade']], x='GENERE')
plt.title('Boxplots of Tests and Average by Gender')
plt.xlabel('Gender')
plt.ylabel('Score')
plt.show()

# 3.3. Boxplots by NATURALESA
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['NATURALESA', 'PCAT', 'PCAST', 'PANG', 'PMAT', 'PMED', 'Average_Grade']], x='NATURALESA')
plt.title('Boxplots of Tests and Average by NATURALESA')
plt.xlabel('NATURALESA')
plt.ylabel('Score')
plt.show()

# 3.4. Boxplots per municipality
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['AREA_TERRITORIAL', 'PCAT', 'PCAST', 'PANG', 'PMAT', 'PMED', 'Average_Grade']], x='AREA_TERRITORIAL')
plt.title('Boxplots of Tests and Average per Municipality')
plt.xlabel('Municipality')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()

# 3.6. Conclusions
print("The box plot analyses of the grades demonstrate significant variation across different ")
print("categories such as year, gender, and nature type. The large boxes and long whiskers indicate")
print("a wide distribution of data and different performance levels within the groups. This suggests")
print("the need for differentiated approaches to teaching and support that can address the diverse needs of students.")
```

## 6. Identification of student types (OPTIONAL)

6\.1. Apply the clustering algorithm k-means on the grades of the 5 subjects. Analyse the results obtained with different numbers of clusters and choose the grouping that you consider most appropriate. Justify the answer. <br>

6\.2. About the chosen solution, explain what type of student belongs to each cluster group. You can use boxplots or histograms discriminating by the cluster identifier (you will have to add the cluster assigned to each row to the student table). 
To save the cluster assigned to each student in the data table, you can do (data is the dataframe that contains the students' data and kmeans the KMeans object):<br>
`dades['cluster'] = kmeans.labels_`

```python
from sklearn.cluster import KMeans

grades = df[['PCAT', 'PCAST', 'PANG', 'PMAT', 'PMED']]

num_clusters = [2, 3, 4, 5, 6]
inertia_values = []

for n in num_clusters:
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(grades)
    inertia_values.append(kmeans.inertia_)

best_num_clusters = num_clusters[inertia_values.index(min(inertia_values))]

kmeans = KMeans(n_clusters=best_num_clusters, random_state=42)
kmeans.fit(grades)


df['cluster'] = kmeans.labels_
print(df.head())
```
